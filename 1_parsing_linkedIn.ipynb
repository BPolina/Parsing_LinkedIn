## Локация вакансии

```python
def get_geo(cell):
    try:
        return BeautifulSoup(cell).find('span', class_ = 'jobs-unified-top-card__bullet').text.strip()
    except:
        return np.nan
```

```python
def get_city(cell):
    if len(cell.split(',')) > 1:
        return cell.split(',')[0].strip()
    elif "Metropolitan" in cell or "Greater" in cell:
        return cell.replace('Greater', '').replace('Metropolitan', '').replace('Area', '').replace('Region','').strip()
    else:
        return np.nan
```

```python
def get_country(cell):
    if len(cell.split(',')) > 1:
        return cell.split(',')[-1].strip()
    elif "Metropolitan" in cell or "Greater" in cell or "Region" in cell:
        return np.nan
    else:
        return cell
```

```python
df['location'] = df['html'].apply(get_geo)
df['city'] = df['location'].apply(get_city)
df['country'] = df['location'].apply(get_country)
```

```python
def pass_and_duplicat (df):
    print('Дубликатов:', df.duplicated().sum())
    print('Пропущенных позиций:')
    print(round((df.isna().mean()*100),2))
```

Необходима дополнительная проработка пропущенных значений. Просмотрим вручную 

```python
loc = {
    'Spain': ['Barcelona', 'Cáceres', 'Gijón', 'Valencia'],
    'Belgium': ['Louvain', 'Antwerp', 'Bruges', 'Brussels', 'Ghent', 'Kortrijk', 'Liege', 'Mons', 'Namur'],
    'France': ['Paris', 'Nantes'],
    'Italy' : ['Turin', 'Barcelona', 'Bologna', 'Genoa', 'Milan', 'Modena', 'Rome'],
    'Poland': ['Gdansk', 'Grudziadz', 'Lodz', 'Przemyśl', 'Radom', 'Warsaw', 'Wroclaw', 'Zamosc'],
    'Germany': ['Berlin'],
    'United Kingdom': ['Edinburgh'],
    'Switzerland':['Zurich'],
    'Sweden': ['Norrköping'],
    'Greece': ['Athens'],
    'Romania':['Bucharest']}
```

```python
def change_the_isna_country (row):
    if str(row['country']) == 'nan':
        city = row['city']
        for k, v in loc.items():
            if city in v:
                return k
    return row['country']
```

```python
df['country'] = df.apply(change_the_isna_country, axis=1)
```

```python
def company_size(size):
    if size == '1-10 employees ':
        return 'microenterprise'
    
    elif size == '11-50 employees ':
        return 'very small company'
    
    elif size == '51-200 employees ':
        return 'small company'
    
    elif size == '201-500 employees ' or size == '201-500 employees':
        return 'medium company'
    
    elif size == '501-1,000 employees ' or size == '1,001-5,000 employees ':
        return 'big company'
    
    elif size == '5,001-10,000 employees ':
        return 'large business'
    
    elif size == '10,001+ employees ':
        return 'corporation'     
    return 'not defined'
```

```python
def get_skills(cell):
    list_skills = []
    for skill in skills:
        if skill in cell.lower().replace('powerbi', 'power bi'):
            list_skills.append(skill)
    return list_skills
```

```python
final_df['skills'] = final_df.description.apply(get_skills)
```
